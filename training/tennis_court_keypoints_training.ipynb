```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms

import json
import cv2
import numpy as np
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
class KeypointsDataset(Dataset):
    def __init__(self, img_dir, data_file):
        self.img_dir = img_dir
        with open(data_file, "r") as f:
            self.data = json.load(f)
        
        self.transforms = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        item = self.data[idx]
        img = cv2.imread(f"{self.img_dir}/{item['id']}.png")
        h,w = img.shape[:2]

        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = self.transforms(img)
        kps = np.array(item['kps']).flatten()
        kps = kps.astype(np.float32)

        kps[::2] *= 224.0 / w # Adjust x coordinates
        kps[1::2] *= 224.0 / h # Adjust y coordinates

        return img, kps
```

```python
train_dataset = KeypointsDataset("data/images","data/data_train.json")
val_dataset = KeypointsDataset("data/images","data/data_val.json")

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=True)
```

```python
model = models.resnet50(pretrained=True)
model.fc =  torch.nn.Linear(model.fc.in_features, 14*2) # Replaces the last layer
```

```python
model = model.to(device)
```

```python
criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

```python
epochs = 20
for epoch in range(epochs):
    model.train()  # Set the model to training mode
    running_loss = 0.0
    
    for i, (imgs, kps) in enumerate(train_loader):
        imgs = imgs.to(device)
        kps = kps.to(device)

        optimizer.zero_grad()
        outputs = model(imgs)  # Shape: [batch_size, 28]
        
        # Ensure keypoints are the same shape as outputs
        kps = kps.view(outputs.shape)  # Reshape if necessary

        loss = criterion(outputs, kps)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        if i % 10 == 0:
            print(f"Epoch [{epoch+1}/{epochs}], Step [{i}/{len(train_loader)}], Loss: {loss.item():.4f}")

    # Print average loss for the epoch
    print(f"Epoch [{epoch+1}/{epochs}], Average Loss: {running_loss / len(train_loader):.4f}")

    # Optional: Validation step after each epoch
    model.eval()  # Set the model to evaluation mode
    val_loss = 0.0
    with torch.no_grad():
        for imgs, kps in val_loader:
            imgs = imgs.to(device)
            kps = kps.to(device)

            outputs = model(imgs)
            kps = kps.view(outputs.shape)  # Ensure shape compatibility

            loss = criterion(outputs, kps)
            val_loss += loss.item()

    print(f"Validation Loss after Epoch [{epoch+1}/{epochs}]: {val_loss / len(val_loader):.4f}")
```

```python
torch.save(model.state_dict(), "keypoints_model.pth")
print("Model checkpoint saved successfully.")
```

```python
def count_parameters(model):
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Total Parameters: {total_params:,}")
    print(f"Trainable Parameters: {trainable_params:,}")

count_parameters(model)
```

```python
def print_gradient_statistics(model):
    for name, param in model.named_parameters():
        if param.grad is not None:
            print(f"{name}: Gradient Norm = {param.grad.norm().item():.6f}")

print("Gradient statistics after training:")
print_gradient_statistics(model)
```

```python
import matplotlib.pyplot as plt

# Visualize a sample prediction
def visualize_prediction(img, true_kps, pred_kps):
    img = img.permute(1, 2, 0).cpu().numpy()  # Convert tensor to numpy image
    plt.imshow(img)
    
    # Plot true keypoints in red
    plt.scatter(true_kps[::2], true_kps[1::2], c='red', label='True Keypoints')
    
    # Plot predicted keypoints in blue
    plt.scatter(pred_kps[::2], pred_kps[1::2], c='blue', label='Predicted Keypoints')
    
    plt.legend()
    plt.show()

# Example usage with a batch from validation loader
imgs, kps = next(iter(val_loader))
imgs, kps = imgs.to(device), kps.to(device)
with torch.no_grad():
    outputs = model(imgs)
    
# Visualize the first image's predictions
visualize_prediction(imgs[0], kps[0].cpu().numpy(), outputs[0].cpu().numpy())
```
